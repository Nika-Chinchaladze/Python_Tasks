```python
try:
  print(5 * 10 - 30 + "25")
except TypeError:
  print("All elements must be Numberic!")
```

ZeroDivisionError Handling:

```python
def Division(a, b):
  try:
    Result = int(a / b)
    return Result
  except ZeroDivisionError:
    ErrorMessage = "Division by zero is forbidden!"
    return ErrorMessage

print(Division(10, 5))
print(Division(5, 0))
print(Division(20, 5))
```

KeyError Handling:

```python
Number_dictionary = {1:10, 2:20, 3:30, 4:40}
try:
  print(Number_dictionary[5])
except KeyError:
  print("Key defined by User, was not found in the dictionary!")
```

Multiple Errors together:

```python
def MultipleErrors(a, b):
  try:
    answer = a / b
    assert a > b
  except ZeroDivisionError:
    answer = "Division by zero is forbidden!"
  except AssertionError:
    answer = "a must be bigger than b"
  return answer

print(MultipleErrors(20, 10))
print(MultipleErrors(20, 0))
print(MultipleErrors(20, 30))
print(MultipleErrors(0, 0))
```

try...except with else:

```python
def try_with_else(a, b):
  try:
    answer = a / b
  except ZeroDivisionError:
    answer = "Division by zero is forbidden!"
  else:
    answer = a * b
  return answer

print(try_with_else(10, 5))
print(try_with_else(10, 0))
print(try_with_else(5, 5))
print("-------------------")
print(try_with_else(10, 10))
print(try_with_else(1, 0))
print(try_with_else(4, 4))
```

try...except_else_finally:

```python
def Return_Boolean():
  try:
    return True
  finally:
    return False
print(Return_Boolean())
```

```python
def Check(a, b):
  try:
    answer = "try: a / b = " + str(a / b)
  except ZeroDivisionError:
    answer = "Division by zero is forbidden!"
  else:
    answer = "else: a * b = " + str(a * b)
  finally:
    answer = "finally: a + b = " + str(a + b)
  return answer

print(Check(10, 5))
print(Check(10, 0))
print(Check(10, 2))
print(Check(5, 3))
```

```python
def Try_Divide(a):
  try:
    answer = a / 0
    print(answer)
  except ZeroDivisionError:
    print("Zero division is forbidden!")
  finally:
    print("This part of the code is always executing!")

Try_Divide(5)
print("-------------")
Try_Divide(3)
```
